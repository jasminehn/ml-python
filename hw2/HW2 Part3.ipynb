```python
import tensorflow
from tensorflow.keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from tensorflow.keras.optimizers import SGD
import numpy as np

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state=99)
```

```python
yTrainRS = np.reshape(y_train, (-1,1))
enc = OneHotEncoder()
enc.fit(yTrainRS)
yTrainENC = enc.transform(yTrainRS).toarray()
yTestRS = np.reshape(y_test, (-1,1))
enc.fit(yTestRS)
yTestENC = enc.transform(yTestRS).toarray()
```

```python
def posOfMaxThree(n1, n2, n3):
    if (n1 >= n2 and n1 >= n3):
        return 0
    if (n2 >= n1 and n2 >= n3):
        return 1
    if (n3 >= n1 and n3 >= n2):
        return 2
```

```python
def predictResults(xTest, yENC):
    results = model.predict(xTest)
    sum = 0
    for i in range(len(yENC)):
        if (posOfMaxThree(yENC[i][0], yENC[i][1], yENC[i][2]) == posOfMaxThree(results[i][0], results[i][1], results[i][2])):
            sum += 1
    print("Accuracy: ", (sum/len(results)))
```

```python
model = Sequential()
model.add(Dense(3, input_dim=4))
model.add(Activation('sigmoid'))
sgd = SGD(learning_rate=0.01)
model.compile(loss='mean_squared_error', optimizer=sgd, metrics = ['accuracy'])
model.fit(X_train, yTrainENC, batch_size=1, epochs=100)
```

```python
model.predict(X_test)
model.evaluate(X_test, yTestENC)
predictResults(X_test, yTestENC)
```
